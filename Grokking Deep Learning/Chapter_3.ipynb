###### simplest neural network possible

```python
# simplest neural network possible

weight = 0.1

def neural_network(input_, weight):
    prediction = input_ * weight
    return prediction
```

```python
# Inserting One Input Datapoint, making the prediction

number_of_toes = [8.5, 9.5, 10, 9]
input_ = number_of_toes[0]
pred = neural_network(input_,weight)
print(pred)
```

So what is a neural network? For now, it's one or
more weights which we can multiply by our input data to make a prediction.

###### Making a Prediction with Multiple Inputs


```python
weights = [0.1, 0.2, 0]

def neural_network(input_, weights):
    pred = w_sum(input_,weights)
    return pred
```

```python
# toes = current number of toes
# wlrec = current games won (percent)
# nfans = fan count (in millions)

toes = [8.5, 9.5, 9.9, 9.0]
wlrec = [0.65, 0.8, 0.8, 0.9]
nfans = [1.2, 1.3, 0.5, 1.0]

input_ = [toes[0],wlrec[0],nfans[0]]
```

```python
def w_sum(a,b):
    assert(len(a) == len(b))
    output = 0
    for i in range(len(a)):
        output += (a[i] * b[i])
    return output
```

```python
pred = neural_network(input_,weights)
print(pred)
```

```python
# numpy code

import numpy as np
weights = np.array([0.1, 0.2, 0])

def neural_network(input_, weights):
    pred = np.dot(input_,weights)
    return pred

toes = np.array([8.5, 9.5, 9.9, 9.0])
wlrec = np.array([0.65, 0.8, 0.8, 0.9])
nfans = np.array([1.2, 1.3, 0.5, 1.0])

# input_ corresponds to every entry
# for the first game of the season
input_ = np.array([toes[0],wlrec[0],nfans[0]])
pred = neural_network(input_,weights)
print(pred)
```
